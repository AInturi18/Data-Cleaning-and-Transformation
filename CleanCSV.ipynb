```python
df = df.drop_duplicates()
print(df)
```

```python
df = df.drop(columns = "Is_Spouse")
```

```python
# df["Accomodation_Type"] = df["Accomodation_Type"].str.strip(",,")
# df["Accomodation_Type"] = df["Accomodation_Type"].str.strip("//")

df["Accomodation_Type"] = df["Accomodation_Type"].str.strip(",/")
df["Holding_Policy_Duration"] = df["Holding_Policy_Duration"].str.strip("+")
df
```

```python
print(df.isnull().sum())
```

```python
import random
df['Health Indicator'] = pd.to_numeric(df['Health Indicator'], errors = 'coerce')
default_value = (['X1','X2','X3','X4','X5','X6'])
df['Health Indicator'] = df['Health Indicator'].apply(lambda x:random.choice(default_value) if pd.isna(x) else x)


df['Holding_Policy_Duration'] = pd.to_numeric(df['Holding_Policy_Duration'], errors = 'coerce')
default_value = df['Holding_Policy_Duration'].median()
df['Holding_Policy_Duration'].fillna(default_value, inplace = True)

df['Holding_Policy_Type'] = pd.to_numeric(df['Holding_Policy_Type'], errors = 'coerce')
default_value = random.randint(1,3)
df['Holding_Policy_Type'].fillna(default_value, inplace = True)
df
```
